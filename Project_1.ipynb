### 3. Time Series Analysis:

#### Analyzing sales trends over time using time series techniques:

```python
# Importing matplotlib and ensuring that the index is set to date:

import matplotlib.pyplot as plt

df.set_index('Date', inplace=True)
df
```

#### 1. What does the Weekly Sales over time for all the stores look like?

```python
plt.figure(figsize=(14, 8))  # Set the size of the plot

# Loop through each store
for store in df['Store'].unique():
    store_data = df[df['Store'] == store]
    plt.plot(store_data.index, store_data['Weekly_Sales'], marker='', linestyle='-', label=f'Store {store}')

plt.title('Weekly Sales Over Time for All Stores')  # Title of the plot
plt.xlabel('Date')  # X-axis label
plt.ylabel('Weekly Sales')  # Y-axis label
plt.legend(title='Stores', bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=len(df['Store'].unique())//5)  # Add legend to the plot
plt.grid(True)  # Add a grid for better readability
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()  # Display the plot
```

#### 2. What is the overall trend in weekly sales for all stores combined?

```python
# Aggregate weekly sales for all stores
overall_sales = df.groupby(df.index)['Weekly_Sales'].sum()

# Plot the overall trend
plt.figure(figsize=(20, 6))
plt.plot(overall_sales)
plt.title('Overall Trend in Weekly Sales')
plt.xlabel('Date')
plt.ylabel('Total Weekly Sales')
plt.grid(True)

# Format x-axis to show each month
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=28))  # Limit number of ticks
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))  # Format dates

plt.xticks(rotation=90)  # Rotate labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

#### 8. Which are the top 3 months of highest sales each year?

```python
# Create 'Year' and 'Month' columns
import calendar

df['Year'] = df.index.year
df['Month'] = df.index.to_series().dt.month.apply(lambda x: calendar.month_abbr[x])


df.head()
```

```python
# Group by 'Year' and 'Month' and calculate total sales for each period
monthly_sales = df.groupby(['Year','Month'])['Weekly_Sales'].sum().reset_index()

top_months_per_year = monthly_sales.sort_values(['Year','Weekly_Sales'], ascending=[True,False])
top_3_months_per_year = top_months_per_year.groupby('Year').head(3)
top_3_months_per_year
```

#### Visualization:

```python
# Group by 'Year' and 'Month' and calculate total sales for each period
monthly_sales = df.groupby(['Year','Month'])['Weekly_Sales'].sum().reset_index()

# Sort by 'Year' and total sales, then get the top 3 months per year
top_months_per_year = monthly_sales.sort_values(['Year','Weekly_Sales'], ascending=[True,False])
top_3_months_per_year = top_months_per_year.groupby('Year').head(3)

# Plotting
plt.figure(figsize=(14, 8))  # Set the size of the plot
for year in top_3_months_per_year['Year'].unique():
    data = top_3_months_per_year[top_3_months_per_year['Year'] == year]
    plt.bar(data['Month'].astype(str) + '-' + data['Year'].astype(str), data['Weekly_Sales'], label=f'Year {year}')

plt.title('Top 3 Months of Highest Sales Each Year')  # Title of the plot
plt.xlabel('Month-Year')  # X-axis label
plt.ylabel('Total Sales')  # Y-axis label
plt.legend(title='Year')  # Add legend to the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for better readability
plt.tight_layout()  # Adjust layout to make room for the plot
plt.show()  # Display the plot
```
